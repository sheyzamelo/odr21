```python
import pandas as pd
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px

# Carregar os dados
file_path = '/content/gdrive/MyDrive/norte.csv'
df = pd.read_csv(file_path, sep='\t')

# Iniciando o aplicativo Dash
app = dash.Dash(__name__)

# Layout do aplicativo
app.layout = html.Div([
    html.H1("Análise de Dados Socioeconômicos por Município na Região Norte do Brasil"),

    html.Label("Selecione a UF:"),
    dcc.Dropdown(
        id='dropdown-uf',
        options=[{'label': uf, 'value': uf} for uf in df['UF'].unique()],
        value='AC'
    ),

    html.Div([
        dcc.Graph(id='scatter-plot'),
        dcc.Graph(id='bar-chart')
    ])
])

# Callback para o gráfico de dispersão
@app.callback(
    Output('scatter-plot', 'figure'),
    [Input('dropdown-uf', 'value')]
)
def update_scatter_plot(selected_uf):
    filtered_df = df[df['UF'] == selected_uf]
    scatter_plot = px.scatter(
        filtered_df,
        x='MUNICÍPIO',
        y='RENDA PER CAPITA CORRIGIDA E AJUSTADA',
        color='TIPOLOGIA SUB REGIONAL',
        title=f'Dispersão da Renda Per Capita em {selected_uf}',
        labels={'RENDA PER CAPITA CORRIGIDA E AJUSTADA': 'Renda Per Capita'}
    )
    return scatter_plot

# Callback para o gráfico de barras
@app.callback(
    Output('bar-chart', 'figure'),
    [Input('dropdown-uf', 'value')]
)
def update_bar_chart(selected_uf):
    bar_chart = px.bar(
        df[df['UF'] == selected_uf],
        x='TIPOLOGIA SUB REGIONAL',
        title=f'Distribuição de Municípios por Tipologia Sub Regional em {selected_uf}',
        labels={'TIPOLOGIA SUB REGIONAL': 'Tipologia Sub Regional'},
        color='TIPOLOGIA SUB REGIONAL'
    )
    return bar_chart

# Executando o aplicativo Dash
if __name__ == '__main__':
    app.run_server(debug=True)
```
